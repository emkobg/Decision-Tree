```python
import pandas as pd #importing pandas  
```

```python
df = pd.read_csv('penguins.csv') #reading in the data
df #displaying data
```

```python
df.columns #displaying names of columns
```

```python
df2 = df[['Island', 'CulmenLength_mm','CulmenDepth_mm', 'FlipperLength_mm', 'BodyMass_g', 'Sex', 'Delta15N','Delta13C']] #feature selection
df2 = df2[df2['Sex'] != '.'] #deleting entries that have "." as entry for Sex since we do not know is it male or female
df2 = df2.dropna() #drops all na values, there were some na values in the Sex column
df2['Sex'].unique() #shows unique elements in 'Sex' columns - just to ensure we onyl have male and female. 
```

```python
df3 = df2 #makes a copy of the dataframe
df3['Sex'] = df2['Sex'].replace('FEMALE',0) #codes females to 0
df3['Sex'] = df2['Sex'].replace('MALE',1) #codes males to 1 
df3 #shows the previous 2 updates 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() #used to create a label encoder
df3['Island'] = le.fit_transform(df3['Island']) #encodes Islands to numbers so they are usable by decision tree
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets, with 80% for training and 20% for testing
train_data, test_data = train_test_split(df3, test_size=0.2, random_state=42) 

# Separate the predictor variables and the variable to predict
predictors_train = train_data.drop('Sex', axis=1)
predictors_test = test_data.drop('Sex', axis=1)
target_train = train_data['Sex']
target_test = test_data['Sex']
```

```python
clf = DecisionTreeClassifier(random_state=20) #creates clf which is a decision tree model
clf.fit(predictors_train, target_train) #trains the model based on training data

# Make predictions on the test data
preds = clf.predict(predictors_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(target_test, preds)
print('Accuracy:', accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf_forrest = RandomForestClassifier(n_estimators=1000, random_state=20) #creates clf_forrest which is a random forrest model
clf_forrest.fit(predictors_train, target_train) #train the random forrest model 

# Make predictions on the test data
preds_forrest = clf_forrest.predict(predictors_test)

# Calculate the accuracy of the model
accuracy_forrest = accuracy_score(target_test, preds_forrest)
print('Accuracy:', accuracy_forrest)
```
